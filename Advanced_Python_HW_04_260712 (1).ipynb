# 01 t-test

수컷과 암컷의 몸무게 차이가 있을까요?

```python
import pandas as pd
```

```python
features = penguins.drop(columns=['species'])
feature_names = features.columns
label = penguins['species']
```

```python
df = pd.DataFrame(features, columns=feature_names)
df['species'] = label
```

```python
male = df[df['sex'] == 'MALE']
female = df[df['sex'] == 'FEMALE']
```

```python
penguins.columns
```

```python
male_weight = male['body_mass_g'].dropna()
```

```python
female_weight = female['body_mass_g'].dropna()
```

```python
from scipy import stats
```

```python
stats.shapiro(male_weight)
```

```python
stats.shapiro(female_weight)
```

```python
stats.levene(male_weight, female_weight)
```

```python
stats.ttest_ind(male_weight, female_weight, equal_var=False)
```

두 집단의 몸무게 평균은 통계적으로 유의미하게 다릅니다.

# 02 ANOVA

펭귄 종에 따라 몸무게가 다를까요?

```python
adelie = df[df['species'] == 'Adelie']
chinstrap = df[df['species'] == 'Chinstrap']
gentoo = df[df['species'] == 'Gentoo']
```

```python
adelie_weight = adelie['body_mass_g'].dropna()
chinstrap_weight = chinstrap['body_mass_g'].dropna()
gentoo_weight = gentoo['body_mass_g'].dropna()
```

```python
import numpy as np
F_statistics, p_val = stats.f_oneway(np.array(adelie_weight),
                                    np.array(chinstrap_weight),
                                    np.array(gentoo_weight))
```

```python
print(f'F statistics is {F_statistics}, p value is {p_val}')
```

세 집단의 평균은 같지 않다.
