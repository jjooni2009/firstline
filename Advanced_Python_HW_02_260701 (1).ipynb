**문제1) 우리는 지난 시간에 setosa의 sepal length에 대해서 하나씩 그래프를 그렸습니다. 다른 iris type의 다른 feature에 대해 똑같이 그래프를 그려볼까요?**

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
feature = iris['data']
```

```python
labels = iris['target']
```

```python
feature_names = iris['feature_names']
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(feature, columns=feature_names)
```

```python
df['target'] = labels
```

```python
df0 = df[df['target'] == 0]
df1 = df[df['target'] == 1]
df2 = df[df['target'] == 2]
```

```python
import matplotlib.pyplot as plt
```

```python
df1['sepal width (cm)']
```

```python
plt.figure(figsize=(6,4))

# 히스토그램을 그려봅시다.
plt.hist(df1['sepal width (cm)'])

# 데이터 표시 없이 깔끔하게 그래프만 딱 그려봅시다.
plt.show()
```

```python
plt.figure(figsize=(6,4))

# line graph를 그려봅시다.
plt.plot(df1['sepal width (cm)'])

# 제목을 써봅시다.
plt.title('Sepal width line graph of iris versicolor', fontsize=14)

# 데이터 표시 없이 깔끔하게 그래프만 딱 그려봅시다.
plt.show()
```

```python
import numpy as np
```

```python
plt.figure(figsize=(6,4))

# bar graph를 그려봅시다.
x = np.arange(len(df1)) # index
plt.bar(x, df1['sepal width (cm)'])
```

```python
plt.figure(figsize=(6,4))

# 도화지의 구성 요소를 쪼갤거예요.
fig, ax = plt.subplots() # 앞으로 ax (axis의 줄임말)에 boxplot을 그릴거예요!!

# boxplot 그려봅시다.
ax.boxplot(df1['sepal width (cm)'])

# 깔끔하게
plt.show()
```

**문제2) 우리는 지난 시간에 모든 iris의 sepal length에 대해서 그래프를 겹쳐 그렸습니다. 다른 feature에 대해 똑같이 그래프를 그려볼까요?**

```python
plt.figure(figsize=(6,4))


plt.hist(df0['sepal width (cm)'], alpha=0.5, label='setosa')
plt.hist(df1['sepal width (cm)'], alpha=0.5, label='versicolor')
plt.hist(df2['sepal width (cm)'], alpha=0.5, label='virginica')


plt.title('Sepal width histograms of three types of iris', fontsize=14)

plt.legend()

plt.show()
```

**문제3) 새로운 와인 데이터를 다뤄봅시다. feature 중 한 가지를 골라서 세 가지의 class를 동시에 그리는 pie chart, box plot, histogram을 그려봅시다. 우리가 그렸던 그래프와 같이 legend도 포함해주세요.**




```python
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
wine
```

```python
feature = wine['data']
```

```python
labels = wine['target']
```

```python
feature_names = wine['feature_names']
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(feature, columns=feature_names)
```

```python
df['target'] = labels
```

```python
df0 = df[df['target'] == 0]
df1 = df[df['target'] == 1]
df2 = df[df['target'] == 2]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,4))


ratio = [len(df0), len(df1), len(df2)]
labels = ['class_0', 'class_1', 'class_2']


plt.pie(ratio, labels=labels, autopct='%.1f%%')

plt.legend(loc='best', bbox_to_anchor=(1,1))


plt.show()
```

```python
plt.figure(figsize=(6,4))

# 도화지의 구성 요소를 쪼갤거예요.
fig, ax = plt.subplots() # 앞으로 ax (axis의 줄임말)에 boxplot을 그릴거예요!!

# boxplot 그려봅시다.
ax.boxplot([df0['alcohol'], df1['alcohol'], df2['alcohol']])

# 제목
plt.title('Alcohol boxplots of three types of wine', fontsize=14)

# 깔끔하게
plt.show()
```

```python
plt.figure(figsize=(6,4))

# 히스토그램을 그려봅시다.
plt.hist(df0['alcohol'], alpha=0.5, label='class_0')
plt.hist(df1['alcohol'], alpha=0.5, label='class_1')
plt.hist(df2['alcohol'], alpha=0.5, label='class_2')

# 제목
plt.title('Alcohol histograms of three types of wine', fontsize=14)

# legend 달기. 꼭 각 색깔에 해당하는 label 정보가 있어야 해요!
plt.legend()

# 데이터 표시 없이 깔끔하게 그래프만 딱 그려봅시다.
plt.show()
```
