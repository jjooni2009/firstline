# 01 Titanic dataset preprocessing
1. 필요 없는 column을 지워보세요.
2. 이번엔 age는 mean으로, Embarked는 median으로 imputation해보세요.
3. Sex, Embarked column을 정수 인코딩 해보세요.

```python
import pandas as pd
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
```

```python
df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
```

```python
df
```

```python
df['Age']
```

```python
df['Age'].mean()
```

```python
df['Age'].fillna(df['Age'].mean())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df
```

```python
df['Embarked'].mode()
```

```python
df['Embarked'].mode()[0]
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df['Embarked']
```

```python
set(df['Embarked'])
```

```python
embarked_mapping = {'C':0, 'Q':1, 'S':2}
df['Embarked'].map(embarked_mapping)
```

```python
df['Embarked'] = df['Embarked'].map(embarked_mapping)
```

```python
mapping = {'male':0, 'female':1}
```

```python
df['Sex'].map(mapping)
```

```python
df['Sex'] = df['Sex'].map(mapping)
```
